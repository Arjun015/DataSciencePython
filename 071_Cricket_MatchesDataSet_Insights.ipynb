## Sorting Data in a series/dataframes

```python
# Based on played number of matches.
```

```python
(dataM['team1'].value_counts() + dataM['team2'].value_counts()).sort_values()
```

```python
# ascending=False => sorting highest to lowest.

(dataM['team1'].value_counts() + dataM['team2'].value_counts()).sort_values(ascending=False)
```

## Dropping the duplicate values

```python
# In a single column
```

```python
dataM.drop_duplicates('season').sort_values('season')
```

```python
dataM.drop_duplicates('season').sort_values('season').shape[0]
```

```python
# Dropping the multiple columns
dataM.drop_duplicates(['season','city']).sort_values(['season','city'],ascending=False)
```

```python
dataM.drop_duplicates('season',keep='last').sort_values('season')
```
