```python
# creating a function to fiter the dataframe by city as argument passed 
```

```python
def match_city(city):
    mask = dataM['city'] == city
    return dataM[mask].shape[0]

match_city('Chennai') #48
match_city('Mumbai')#85
match_city('Bangalore')#66
    
```

```python
def match_city(city):
    mask = dataM['city'] == city
    result =  dataM[mask].shape[0]
    print(f'The No of matches played in {city} is {result}')

match_city('Chennai') 
match_city('Mumbai')#85
match_city('Bangalore')#66
```
