```python
# PUBLIC VARIABLE Encapsulation.

# 1.Public variable encapsulation is a type of encapsulation in Python
# 2.where the data members of a class are public, 
# 3.which means they can be accessed from anywhere within the program. 
# 4.This type of encapsulation is not as secure as private variable encapsulation, 
# 5.but it can be useful in some cases.


class Account :
    counter = 0
    def __init__(self,opening_balance=0):
        Account.counter = Account.counter + 1
        self.id = Account.counter
        self.balance = opening_balance
        self.num_tx = 0
        self.max_tx = 5
        
    def __str__(self):
        return f"Account number is: {self.id} and Balance is :{self.balance}"
    
    
    def deposite(self,amount):
        if amount > 0 and self.num_tx < self.max_tx :
            self.balance = self.balance + amount
            self.num_tx  = self.num_tx + 1
        
    def withdrawl(self,amount):
        if amount > 0 and self.balance >= amount and self.num_tx < self.max_tx:
            self.balance = self.balance - amount
            self.num_tx = self.num_tx + 1
            
        
        

class SavingAccount(Account):
    pass

class CurrentAccount(Account):
    def __init__(self):
        super().__init__()
        self.max_tx = 100  
sa1 = SavingAccount()
ca1 = CurrentAccount()

print(sa1)
print(ca1)


sa1.balance = 10000
print(sa1)
```

```python
# PRIVATE VARIABLE Encapsulation.

# 1.Private variable encapsulation is a type of encapsulation in Python.
# 2.where the data members of a class are private, 
# 3.which means they can only be accessed by methods within the class. 
# 4.This type of encapsulation is more secure than public variable encapsulation, 
# 5.because it prevents other objects from directly modifying the value of the data members.

# hiding attribute using Name Mangling
#is a technique that can be used to further protect private attributes in Python.

class Account :
    counter = 0
    def __init__(self,opening_balance=0):
        Account.counter = Account.counter + 1
        self.id = Account.counter
        self.__balance = opening_balance
        self.num_tx = 0
        self.max_tx = 5
        
    def __str__(self):
        return f"Account number is: {self.id} and Balance is :{self.__balance}"
    
    
    def deposite(self,amount):
        if amount > 0 and self.num_tx < self.max_tx :
            self.__balance = self.__balance + amount
            self.num_tx  = self.num_tx + 1
        
    def withdrawl(self,amount):
        if amount > 0 and self.__balance >= amount and self.num_tx < self.max_tx:
            self.__balance = self.__balance - amount
            self.num_tx = self.num_tx + 1
            
        
        

class SavingAccount(Account):
    pass

class CurrentAccount(Account):
    def __init__(self):
        super().__init__()
        self.max_tx = 100  
        
        
sa1 = SavingAccount()
ca1 = CurrentAccount()

print(sa1)
print(ca1)

sa1.__balance = 10000
print(sa1.__balance)

ca1.__balance = 20000
print(ca1.__balance)


# Name Mangling.

print(sa1._Account__balance)
# person = Person()
# print(person._Person__name)

```

```python
# PROTECTED VARIABLE Encapsulation.
# THIS IS SYMBOLIC IN NATURE.

# 1.Protected variable encapsulation is a type of encapsulation in Python.
# 2.where the data members of a class are protected, 
# 3.which means they can only be accessed by methods within the class and by subclasses of the class. 
# 4.This type of encapsulation is less secure than private variable encapsulation, 
# 5.but it is more flexible.


class Account :
    counter = 0
    def __init__(self,opening_balance=0):
        Account.counter = Account.counter + 1
        self.id = Account.counter
        self._balance = opening_balance
        self.num_tx = 0
        self.max_tx = 5
        
    def __str__(self):
        return f"Account number is: {self.id} and Balance is :{self._balance}"
    
    
    def deposite(self,amount):
        if amount > 0 and self.num_tx < self.max_tx :
            self._balance = self._balance + amount
            self.num_tx  = self.num_tx + 1
        
    def withdrawl(self,amount):
        if amount > 0 and self._balance >= amount and self.num_tx < self.max_tx:
            self._balance = self._balance - amount
            self.num_tx = self.num_tx + 1
            
        
        

class SavingAccount(Account):
    pass

class CurrentAccount(Account):
    def __init__(self):
        super().__init__()
        self.max_tx = 100  
        
        
sa1 = SavingAccount()
ca1 = CurrentAccount()

print(sa1)
print(ca1)

sa1._balance = 40000
print(sa1._balance)

sa1._balance = 20000
print(sa1._balance)
```
